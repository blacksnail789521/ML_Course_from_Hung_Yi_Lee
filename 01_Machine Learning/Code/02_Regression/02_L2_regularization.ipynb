```python
import numpy as np
from sklearn.linear_model import Ridge # Linear least squares with l2 regularization.
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {"axes.titlesize": 20, "axes.labelsize": 15, "lines.linewidth": 3})
```

```python
# Create data from -1 to 1.
np.random.seed(0)
number_of_samples = 100
X = np.random.rand(number_of_samples, 1) * 2 - 1
y = np.random.rand(number_of_samples) * 2 - 1
```

```python
# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 1)
```

```python
def get_mse(X, y):
    y_pred = model.predict(X)
    y_mse = mean_squared_error(y, y_pred)
    
    return y_mse


alpha_list = list(np.logspace(0, 3, 100))
training_mse_list = []
testing_mse_list = []
for alpha in alpha_list:
    
    # Train model with linear regression.
    model = Ridge(alpha).fit(X_train, y_train)
    
    # Get mse for both training set and testing set.
    training_mse_list.append( get_mse(X_train, y_train) )
    testing_mse_list.append( get_mse(X_test, y_test) )

# Plot the result.
plt.plot(alpha_list, training_mse_list, color = "red",  label = "training")
plt.plot(alpha_list, testing_mse_list,  color = "blue", label = "testing")
plt.legend()
plt.xscale("log")
plt.xlabel("Level of regularization")
plt.ylabel("Performance (MSE)")
plt.title("Regularization")
plt.show()
```
