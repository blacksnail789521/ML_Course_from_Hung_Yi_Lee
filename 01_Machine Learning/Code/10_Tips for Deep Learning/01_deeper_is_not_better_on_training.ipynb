```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {"axes.titlesize": 20, "axes.labelsize": 15, "legend.fontsize": 15, \
              "lines.linewidth": 3, "figure.figsize": (9, 4)})
```

```python
import tensorflow as tf

# Set parameters of data.
num_classes = 10
input_shape = (28 * 28,)

# Get training_data and testing_data.
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Make sure each image has the shape (28 * 28, 1).
x_train = x_train.reshape(x_train.shape[0], 28 * 28)
x_test = x_test.reshape(x_test.shape[0], 28 * 28)

# Scale images to the [0, 1] range.
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
def build_model(num_hidden_layer):

    # Build the model.
    model = tf.keras.Sequential(name = "DNN")
    model.add( tf.keras.Input(shape=input_shape) )
    for i in range(num_hidden_layer):
        model.add( tf.keras.layers.Dense(500, activation="sigmoid", name=f"hidden_layer_{i+1}") )
    model.add( tf.keras.layers.Dense(num_classes, activation="softmax", name="output_layer") )
    
    return model
```

```python
def train_model(model):

    # Train the model.
    batch_size = 128
    epochs = 15

    model.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])

    model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1, verbose=0)
    
    return model
```

```python
def evaluate_trained_model(model):
    
    # Evaluate the trained model.
    training_loss, training_accuracy = model.evaluate(x_train, y_train, verbose=0)
    testing_loss, testing_accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    return training_accuracy, testing_accuracy
```

```python
accuracy_dict = { num_hidden_layer: np.nan for num_hidden_layer in range(10) }
for num_hidden_layer in tqdm(accuracy_dict.keys()):
    
    model = build_model(num_hidden_layer)
    model = train_model(model)
    training_accuracy, testing_accuracy = evaluate_trained_model(model)
    
    # Store test_accuracy into accuracy_dict.
    accuracy_dict[num_hidden_layer] = {"training_accuracy": training_accuracy * 100, \
                                       "testing_accuracy": testing_accuracy * 100}
```

```python
num_hidden_layer_list = accuracy_dict.keys()
training_accuracy_list = [ accuracy["training_accuracy"] for num_hidden_layer, accuracy in accuracy_dict.items() ]
testing_accuracy_list = [ accuracy["testing_accuracy"] for num_hidden_layer, accuracy in accuracy_dict.items() ]


plt.plot(num_hidden_layer_list, training_accuracy_list, color = "red",  label = "training")
plt.plot(num_hidden_layer_list, testing_accuracy_list,  color = "blue", label = "testing")

plt.xlabel("Layers")
plt.ylabel("Accuracy (%)")
plt.title("Handwritten Digit Classification")
plt.legend()
plt.show()
```
