```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set(rc = {"axes.titlesize": 20, "axes.labelsize": 15, "legend.fontsize": 15, \
#               "lines.linewidth": 3, "figure.figsize": (9, 4)})
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

def create_dataset():
    
    X, y = make_moons(noise=0.3, random_state=0)
    X = StandardScaler().fit_transform(X)

    # Split data.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
    
    return X_train, X_test, y_train, y_test


X_train, X_test, y_train, y_test = create_dataset()
```

```python
from sklearn.neural_network import MLPClassifier

def train_model(X_train, y_train):
    
    # Fit the classifier
    clf = MLPClassifier(solver='lbfgs', random_state=1, max_iter=2000).fit(X_train, y_train)
    
    return clf
    

clf = train_model(X_train, y_train)
```

```python
from matplotlib.colors import ListedColormap

def plot_result(X_train, X_test, y_train, y_test, clf):
    
    # Plot initialization.
    plt.figure(figsize=(17, 5))
    x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
    y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, .02),
                         np.arange(y_min, y_max, .02))
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
    Z = Z.reshape(xx.shape)

    # Plot both training set and testing set without decision boundary.
    plt.subplot(1, 3, 1)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, \
                alpha=0.15, marker="^")
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("Training Set + Testing Set")

    # Plot training set with decision boundary.
    plt.subplot(1, 3, 2)
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=.8)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, \
                edgecolors='black')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    score = clf.score(X_train, y_train)
    plt.title(f"Training Set (score: {score})")

    # Plot testing set with decision boundary.
    plt.subplot(1, 3, 3)
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=.8)
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, \
                edgecolors='black', marker="^")
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    score = clf.score(X_test, y_test)
    plt.title(f"Testing Set (score: {score})")

    plt.show()
    
# Show result.
plot_result(X_train, X_test, y_train, y_test, clf)
```
